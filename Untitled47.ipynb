```python
#armstrong
def armstrong(num):
    strings=str(num)
    n=len(strings)

    armstrong=sum(int(string)**n for string in strings)
    if armstrong==num:
        print("armstrong")
    else:
        return ("not armstrong")


```

```python
#matrix addition
def matrix_addition(matrix1,matrix2):
    result=[[matrix1[i][j]+matrix2[i][j] for j in range(len(matrix1[0]))]for i in range(len(matrix1))]
    return result
```

```python
matrix1 = [[1, 2, 3]]
matrix2 = [[4, 5, 6]]
matrix_addition(matrix1, matrix2)
```

```python
#matrix multiplication
def matrix_multiplication(matrix1, matrix2):
    # Get dimensions
    rows_matrix1 = len(matrix1)
    cols_matrix1 = len(matrix1[0])
    cols_matrix2 = len(matrix2[0])

    # Initialize result matrix with zeros
    result = [[0 for _ in range(cols_matrix2)] for _ in range(rows_matrix1)]

    # Perform multiplication
    for i in range(rows_matrix1):
        for j in range(cols_matrix2):
            for k in range(cols_matrix1):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    return result
```

```python
matrix1 = [
    [1, 2, 3],
    [4, 5, 6],
]

matrix2 = [
    [7, 8],
    [9, 10],
    [11, 12]
]
matrix_multiplication(matrix1, matrix2)
```

```python
#lcm of 2 numbers
def calculate_lcm(a, b):
    # Find the LCM using the formula: LCM(a, b) = abs(a * b) // GCD(a, b)
    def gcd(x, y):
        while y:
            x, y = y, x % y
        return x

    return abs(a * b) // gcd(a, b)

# Input two numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Calculate and display LCM
lcm = calculate_lcm(num1, num2)
print(f"The LCM of {num1} and {num2} is: {lcm}")
```

```python
#standardizze
std=(new_df-(np.mean(new_df)))/(np.std(new_df))
print(std)
```

```python
#covariance matrix
cov_matrix=np.cov(std,rowvar=False)
print(cov_matrix)
```

```python
#eigen value and eigen vectors
eigenvalues,eigenvectors=np.linalg.eigh(cov_matrix)
print(eigenvalues)
print(eigenvectors)
```

```python
#sort
sorted_indices=np.argsort(eigenvalues)[::-1]
sorted_eigenvalues=eigenvalues[sorted_indices]
sorted_eigenvectors=eigenvectors[:,sorted_indices]
```

```python
#principle component
n=int(input("Enter the number of PCA:"))
selected_eigenvectors=sorted_eigenvectors[:,:n]
```

```python
pca_data=np.dot(std,selected_eigenvectors)
print(pca_data)
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
y = iris.target

# Specify the unknown tuple
point = np.array([6.5, 3.0, 5.5, 1.8])

# Calculate distances using Euclidean distance for each data point
distances = np.sqrt(np.sum((df.values - point) ** 2, axis=1))

# Value of k
k = int(input('Enter the value of k: '))

# Find indices of the k-nearest neighbors for each data point
k_indices = np.argsort(distances)[:k]

# Get labels for the k-nearest neighbors for each data point
k_labels = y[k_indices]
iris.target_names[k_labels]

# Determine the majority class among the k-nearest neighbors for each data point
predictions = max(set(k_labels), key=list(k_labels).count)
print(iris.target_names[predictions])
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

# Load the iris dataset
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Manage the data
X = data[['sepal length (cm)']]  # Ensure the correct column name is used
Y = data['petal width (cm)']  # Ensure the correct column name is used

# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# Train the model
model = SVR(kernel='linear')
model.fit(X_train, Y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(Y_test, y_pred)
print("Mean Squared Error:", mse)
```
