```python
#write a function that can return the track record of 2 teams against each other
def matches(a,b):
    mask1=ipl['Team1']==a
    mask2=ipl['Team2']==b
    mask3=ipl['Team1']==b
    mask4=ipl['Team2']==a

    relevant_matches = ipl[mask1 & mask2 | mask3 & mask4]#or function

    # Calculate wins for Team 'a' and Team 'b'
    team_a_wins = relevant_matches[(relevant_matches['WinningTeam'] == a)].shape[0]
    team_b_wins = relevant_matches[(relevant_matches['WinningTeam'] == b)].shape[0]

    # Return the counts
    return f"Matches Played: {relevant_matches.shape[0]}, {a} Won: {team_a_wins}, {b} Won: {team_b_wins}"
```

```python
def euclidian(row):
    pt_A=row['1st point']
    pt_B=row['2nd point']

    return ((pt_A[0]-pt_B[0])**2 +(pt_A[1]-pt_B[1])**2)**0.5
```

```python
#FIND normalized imdb rating grp wise
def normal(group):
    rating_range = group['IMDB_Rating'].max() - group['IMDB_Rating'].min()
    if rating_range == 0:
        group['norm_equal'] = 0
    else:
        group['norm_equal'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / rating_range
    return group
```
