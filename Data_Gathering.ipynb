```python
#all the code in single snippet for a singl car

car_name = soup.find('h1').text.strip()

# Extract rating
rating = soup.find('span', class_='ratingStarNew').text.strip()

# Extract price
price = soup.find('div', class_='price').text.strip()

# Extract specifications from the table
specifications = []
rows = soup.find_all("tbody")[0].find_all('tr')
for row in rows:
    td_text = row.find('td').text.strip()  # First <td> text
    span_text = row.find('span').text.strip()  # <span> text
    specifications.append(f"{td_text}: {span_text}")

# Print the extracted information
print(f"Car Name: {car_name}")
print(f"Rating: {rating}")
print(f"Price: {price}")
print("Specifications:")
for spec in specifications:
    print(f"  - {spec}")
```

```python

import pandas as pd
# List of URLs for different cars
urls = [
    "https://www.cardekho.com/tata/curvv-ev/specs",
    "https://www.cardekho.com/mahindra/be-6/specs",
    "https://www.cardekho.com/tata/nexon-ev/specs",
    "https://www.cardekho.com/tata/punch-ev/specs",
    "https://www.cardekho.com/tata/tiago-ev/specs",
    "https://www.cardekho.com/mg/comet-ev/specs",
    "https://www.cardekho.com/mahindra/xev-9e/specs",
    "https://www.cardekho.com/byd/emax-7/specs",
    "https://www.cardekho.com/byd/atto-3/specs",
    "https://www.cardekho.com/mg/zs-ev/specs",
    "https://www.cardekho.com/mg/windsor-ev/specs",
    "https://www.cardekho.com/mahindra/xuv400-ev/specs",

]

cars_data=[]
# Loop through each URL
for url in urls:
    print(f"Fetching data for: {url}")

    # Fetch the HTML content of the page
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'})

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract car name
        car_name = soup.find('h1').text.strip() if soup.find('h1') else "N/A"

        # Extract rating
        rating = soup.find('span', class_='ratingStarNew').text.strip() if soup.find('span', class_='ratingStarNew') else "N/A"

        # Extract price
        price = soup.find('div', class_='price').text.strip() if soup.find('div', class_='price') else "N/A"

        # Extract specifications
        specifications = {}
        tbody = soup.find_all("tbody")
        if tbody:
            rows = tbody[0].find_all('tr')
            for row in rows:
                td_text = row.find('td').text.strip() if row.find('td') else "N/A"
                span_text = row.find('span').text.strip() if row.find('span') else "N/A"
                specifications[td_text] = span_text

        # Combine all data for the current car
        car_data = {
            "Car Name": car_name,
            "Rating": rating,
            "Price": price,
            **specifications  # Add specifications as additional columns
        }

        # Append the data to the list
        cars_data.append(car_data)
    else:
        print(f"Failed to fetch data for {url}. Status Code: {response.status_code}")

# Create a DataFrame from the list of car data
df = pd.DataFrame(cars_data)

# Display the DataFrame

```
